## Libraries to Load

```r
library(tibble)
library(ggplot2)
```

## Random Data
The following is random data generated to test the code.  Actual data will be expected in form of csv file and uploaded in the same directory as the Jupyter notebook.

```r
rnorm(100, mean=7, sd=1) -> weeksleep
rnorm(100, mean=8, sd=1) -> satsleep
```

```r
weeklabel <- rep("week", length(weeksleep))
satlabel <- rep("sat", length(satsleep))
```

```r
type <- c(weeklabel, satlabel)
hours <- c(weeksleep, satsleep)
```

```r
df <- tibble(
    type, hours)
```
